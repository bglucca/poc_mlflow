```python
zero_cols = ['price',
             'sales',
             'fba_fees',
             'active_sellers',
             'ratings',
             'images',
             'review_count']

median_cols = ['ratings',
               'review_velocity',
               'bsr']
```

```python
for col in zero_cols:

    df_raw[col].fillna(0, inplace = True)

for col in median_cols:

    median = df_raw[col].median()

    df_raw[col].fillna(median, inplace = True)

    if col == 'review_velocity':

        df_raw.loc[df_raw[col] < 0, col] = median

    if col == 'bsr':
        
        df_raw.loc[df_raw[col] == 0, col] = median
```

```python
df_raw['bsr'].fillna(5e6, inplace=True)
```

# Categories

```python
cat_cols = df_raw.select_dtypes(include='object').columns.drop('asin')
```

```python
for col in cat_cols:

    print(f'Unique values in {col}:', df_raw[col].nunique())
```

Category might be ok to leave as such, but there are too many brands. Let's bin this category into top-selling brands

```python
top_selling_brands = df_raw.groupby('brand', as_index = False)['sales'].sum()

top_selling_brands['prop'] = (top_selling_brands['sales'] / top_selling_brands['sales'].sum()).round(4)

top_selling_brands

top_selling_brands.sort_values('prop', ascending = False, inplace = True)

top_selling_brands['prop_acum'] = top_selling_brands['prop'].cumsum()

top_selling_brands
```

```python
top_sellers = top_selling_brands[top_selling_brands['prop_acum'] <= 0.95]['brand'].values
```

```python
df_raw['top_selling_brand'] = df_raw['brand'].isin(top_sellers)
```
